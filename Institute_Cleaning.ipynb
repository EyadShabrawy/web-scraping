```python
import pandas as pd
df = pd.read_csv('Institute.csv')
```

```python
df
```

```python
df = df[['Name', 'address', 'website']].head(10)
```

```python
df
```

```python
df['Name'] = df['Name'].str.split('-').str[0]
```

```python
df = df.rename(columns={
    'address': 'Address',
    'website': 'Contact'
})
```

```python
df.index += 1
df.reset_index(inplace=True)
df.rename(columns={'index': 'InstituteID'}, inplace=True)
```

```python
df
```

```python
df.to_csv("Institute.csv", index = False)
```
