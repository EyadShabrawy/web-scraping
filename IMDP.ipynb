```python
import pandas as pd

df = pd.read_html('https://www.imdb.com/chart/top/')
```

```python
df = df[0]
df
```

```python
df.drop(columns=["Unnamed: 0", "Unnamed: 4", "Your Rating"], inplace = True)
```

```python
df
```

```python
df[ ["Rank","Title"] ] = df[ "Rank & Title" ].str.split(".", expand=True, n = 1)
```

```python
df
```

```python
df.drop(columns=["Rank & Title"], inplace = True)
```

```python
df
```

```python
df[["Title", "Year"]] = df["Title"].str.split("(", expand=True)
```

```python
df["Year"] = df["Year"].str.replace(")", "")
```

```python
df
```

```python
custom_order=["IMDb Rating", "Year", "Rank", "Title"]
df = df.reindex(columns=custom_order)
```
