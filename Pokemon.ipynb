```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome import service

import time
```

```python
url="https://pokemondb.net/pokedex/all"
```

```python
webdriver_service = service.Service("/usr/local/bin/chromedriver.exe")
```

```python
browser = webdriver.Chrome(service = webdriver_service)
browser.get(url)
```

```python
id_objects = browser.find_elements(By.CLASS_NAME, "infocard-cell-data")
id_list= [ ]
for id in id_objects:
    id_list.append(id.text)
```

```python
id_list
```

```python
names_objects = browser.find_elements(By.CLASS_NAME, "ent-name")
names_list= [ ]
for name in names_objects:
    names_list.append(name.text)
```

```python
names_list
```

```python
type_objects = browser.find_elements(By.CLASS_NAME, "cell-icon")
type_list= [ ]
for type in type_objects:
    type_list.append(type.text)
```

```python
type_list
```

```python
total_objects = browser.find_elements(By.CLASS_NAME, "cell-num.cell-total")
total_list= [ ]
for total in total_objects:
    total_list.append(total.text)
```

```python
total_list
```

```python
stats_objects = browser.find_elements(By.CLASS_NAME, "cell-num")
stats_list= [ ]
for stat in stats_objects:
    stats_list.append(stat.text)
```

```python
stats_list
```

```python
hp = []
attack = []
defense = []
sp_atk = []
sp_def = []
speed = []
```

```python
for i in range(len(stats_list)):
    if i % 8 == 2:
        hp.append(stats_list[i])
    elif i % 8 == 3:
        attack.append(stats_list[i])
    elif i % 8 == 4:
        defense.append(stats_list[i])
    elif i % 8 == 5:
        sp_atk.append(stats_list[i])
    elif i % 8 == 6:
        sp_def.append(stats_list[i])
    elif i % 8 == 7:
        speed.append(stats_list[i])
```

```python
import pandas as pd
data = {
    "IDs": id_list,
    "Names": names_list,
    "Type": type_list,
    "Total": total_list,
    "HP": hp,
    "Attack": attack,
    "Defense": defense,
    "Sp. Atk": sp_atk,
    "Sp. Def": sp_def,
    "Speed": speed,
}
```

```python
df = pd.DataFrame.from_dict(data)
```

```python
df
```

```python
df["Type"] = df["Type"].apply(lambda s: s.replace("\n", " & ") )
```

```python
df
```

```python
df.to_excel("pokemon_dataset.xlsx", index = False)
```
